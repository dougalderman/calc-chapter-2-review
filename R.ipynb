```r
x<-c(0,27,32,37,42,47,52,57,62,67,72,77,82,87)
y<-c(0,10,25,60,125,187,224,270,340,408,437,475,460,420)
```

```r
plot(x,y,col="blue", main="Incidence of Breast Cancer Versus Age",xlab="Age", ylab="Breast Cancer Incidence Per 100,000")
```

```r
relation<-lm(y~x)
print(relation)
summary(relation)
```

```r
x_seq <- seq(from=0,to=90, by=0.1)
new_relation<-list(0)
for (i in 1:length(x_seq)) {
    x_new <- data.frame(x=x_seq[i])
    new_relation[i]<-predict(relation, x_new)
}
plot(x,y,col="blue", main="Linear Function",lines(x_seq, new_relation),xlab="x",ylab="y")
```

```r
confint(relation, level=0.95)
```

```r
plot(fitted(relation), residuals(relation))
```

```r
relation2<-lm(y~x + I(x^2))
print(relation2)
summary(relation2)
```

```r
x_seq <- seq(from=0,to=90, by=0.1)
new_relation2<-list(0)
for (i in 1:length(x_seq)) {
    x_new <- data.frame(x=x_seq[i])
    new_relation2[i]<-predict(relation2, x_new)
}
plot(x,y,col="blue", main="Quadtratic Function",lines(x_seq, new_relation2),xlab="x",ylab="y")
```

```r
confint(relation2, level=0.95)
```

```r
plot(fitted(relation2), residuals(relation2))
```
